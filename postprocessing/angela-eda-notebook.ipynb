### Yearly breakdown of Legacy scraped data by age, race and gender

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filepath = 'June_2022_postprocessed/'
all_dfs = []
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021]
for year in years:
    infile = pd.read_csv(filepath + '{}_postprocessed.csv'.format(year))
    all_dfs.append(infile)
```

```python
## Output: dataframe with yearly break-down of given demographic variable
## Input: all_dfs (lists), variable(str)##
def get_yearly_counts(all_dfs, variable):
    df = pd.DataFrame()
    for i in range(len(all_dfs)):
        year_df = all_dfs[i]
        df[str(years[i])] = year_df.groupby(variable).count()['name']
    df.fillna(0, inplace=True)
    return df
```

```python
df = get_yearly_counts(all_dfs, 'pred_race')
```

```python
def plot_stacked_bars(df):
    df.transpose().plot(kind='bar', stacked=True)
    percentage_df = df/df.sum()
    percentage_df.transpose().plot(kind='bar', stacked=True)

plot_stacked_bars(df)
    
```
